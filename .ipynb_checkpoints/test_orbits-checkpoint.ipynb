Import needed packages

```python
import numpy as np
import matplotlib.pylab as plt
import pyproj

from PyAstronomy import pyasl
from mpl_toolkits.mplot3d import Axes3D
```

Create Keplerian orbit elements for DSX and VPM

```python
# Instantiate Keplerian elliptical orbits with
# semi-major axis [m], period [sec], eccentricity, and inclination [deg]

mu = 3.986004418e14 # [m^3/s^2] gravitational parameter for Earth
R_E = 6.371e6       # [m] Earth radius
dsx_a = 9e6 + R_E   # [m] semi major axis
vpm_a = 5e5 + R_E   # [m] semi major axis

dsx_period = 2*np.pi*np.sqrt(dsx_a**3/mu)
vpm_period = 2*np.pi*np.sqrt(vpm_a**3/mu)

dsx = pyasl.KeplerEllipse(dsx_a, dsx_period, e=0.33333, i=120)
vpm = pyasl.KeplerEllipse(vpm_a, vpm_period, e=0, i=51.6)

# Get a time axis
t = np.linspace(0, dsx_period*1.25, 200)
```

We now have arrays of positions from the center of the earth that we can plot

```python
# Calculate the orbit position at the given points
# in a Cartesian coordinate system

# initiate 3D axes
ax = Axes3D(plt.gcf())

# draw Earth sphere
u = np.linspace(0, 2 * np.pi, 200)
v = np.linspace(0, np.pi, 200)
x = R_E*np.outer(np.cos(u), np.sin(v))
y = R_E*np.outer(np.sin(u), np.sin(v))
z = R_E*np.outer(np.ones(np.size(u)), np.cos(v))

ax.plot_surface(x, y, z, linewidth=0.0,color = 'g')

# get ECEF coordinates
dsx_pos = dsx.xyzPos(t)
vpm_pos = vpm.xyzPos(t)

dsx_x = dsx_pos[::,0]
dsx_y = dsx_pos[::,1]
dsx_z = dsx_pos[::,2]
vpm_x = vpm_pos[::,0]
vpm_y = vpm_pos[::,1]
vpm_z = vpm_pos[::,2]

ax.scatter( dsx_x, dsx_y, dsx_z)
ax.scatter( vpm_x, vpm_y, vpm_z )

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_xlim3d(-dsx_a*1.5,dsx_a*1.5)
ax.set_ylim3d(-dsx_a*1.5,dsx_a*1.5)
ax.set_zlim3d(-dsx_a*1.5,dsx_a*1.5)
ax.set_title('DSX and VPM Orbits')
```

Let's convert to ECEF and then to LLA so we can get L-shell and latitudes of the spacecrafts. 

```python
ecef = pyproj.Proj(proj='geocent', ellps='WGS84', datum='WGS84')
lla = pyproj.Proj(proj='latlong', ellps='WGS84', datum='WGS84')

dsx_lat, dsx_lon, dsx_alt = pyproj.transform(ecef, lla, dsx_x, dsx_y, dsx_z, radians=False)
vpm_lat, vpm_lon, vpm_alt = pyproj.transform(ecef, lla, vpm_x, vpm_y, vpm_z, radians=False)
```

Now let's grab a time and see where the two are at

```python
# at timestep k
k = 100
print('DSX is at ',dsx_lat[k],dsx_lon[k],dsx_alt[k])
print('VPM is at ',vpm_lat[k],vpm_lon[k],vpm_alt[k])
```
